## REMOVE DUPLICATE

```python
df_lazada.dropna(axis=1)
df_lazada.drop_duplicates(inplace=True)
df_lazada.head()
```

## CLEANSING

```python
def cleaning(data):
    data = re.sub(r"\n", " ",data) # menghilangkan tanda baca
    data = re.sub(r"\r", " ",data)
    data = re.sub(r"&amp;", " ",data)
    data = re.sub(r"&gt;", " ",data)
    data = re.sub(r"&lt;", " ",data)
    data = re.sub(r"&quot;", " ",data)
    data = re.sub(r"\u00a0", " ",data)
    data = re.sub(r'''(?i)\b((?:https|http?://|www\d{0,3}[.]|[a-z0-9.\-]+[.][a-z]{2,4}/)(?:[^\s()<>]+|\(([^\s()<>]+|(\([^\s()<>]+\)))*\))+(?:\(([^\s()<>]+|(\([^\s()<>]+\)))*\)|[^\s`!()\[\]{};:'".,<>?«»“”‘’]))''', "",data)
    data = re.sub(r"'s","",data)
    data = re.sub(r"can't","can not",data) # mengubah kata
    data = re.sub(r"ain't","am not",data)
    data = re.sub(r"won't","will not",data)
    data = re.sub(r"n't"," not",data)
    data = re.sub(r"'re"," are",data)
    data = re.sub(r"[-_<>*#$@()\"/%&;:<>{}`+=~|.!?,']", " ", data) #hapus simbol
    data = re.sub(r"[^a-zA-Z]", " ",data) # menghilangkan angka
    data = re.sub("\s\s+", " ",data) # menghilangkan double spasi
    return data
```

```python
# menambahkan kolom baru yang berisikan reviewText yang sudah bersih dari unwanted character
df_lazada['cleansing'] = df_lazada['review'].apply(lambda x: cleaning(x))
df_lazada.iloc[0:800]
```

## CASE FOLDING

```python
df_lazada['case_folding'] = df_lazada['cleansing'].str.lower()
df_lazada.loc[0:300,['cleansing', 'case_folding']]
```

```python
# menambahkan kolom baru yang berisikan reviewText yang sudah bersih dari unwanted character
df_lazada['no_unwanted'] = df_lazada['case_folding'].apply(lambda x: cleaning(x))
df_lazada.head()
```

##TOKENIZATION


```python
# # NLTK word rokenize

nltk.download('punkt')
def tokenize(text):
    return word_tokenize(text);

df_lazada['tokenize'] = df_lazada['case_folding'].apply(tokenize);
df_lazada.loc[0:300,['case_folding', 'tokenize']]
```

##STEMMING

```python
# !pip install Sastrawi
```

```python
# !pip install swifter
```

```python
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import swifter


# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# stemmed
def stemmed_wrapper(term):
    return stemmer.stem(term)

term_dict = {}

for document in df_lazada['tokenize']:
    for term in document:
        if term not in term_dict:
            term_dict[term] = ' '

for term in term_dict:
    term_dict[term] = stemmed_wrapper(term)

# apply stemmed term to dataframe
def get_stemmed_term(document):
    return [term_dict[term] for term in document]

df_lazada['stemming'] = df_lazada['tokenize'].swifter.apply(get_stemmed_term)

df_lazada.loc[0:300,['tokenize', 'stemming']]
```

## Preprocessing Result

```python
df_lazada["clean"] = df_lazada["stemming"].apply(lambda x: " ".join(x))
df_lazada.head()
```
